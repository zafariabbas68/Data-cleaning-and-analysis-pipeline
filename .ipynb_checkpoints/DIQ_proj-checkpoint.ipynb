Let us have a look at STREET.

We can find a street from its coordinates in different ways. We tried automated street localization from coordinates, but the queries took way too long too run. Here we simply take the closest point that has an associated street, and retrieve this street, for all points without a street.

```python
points_without_street = df[(df.STREET.isnull()) & (df.Lat.notnull()) & (df.Long.notnull())][['Lat', 'Long']]
points_with_street = df[(df.STREET.notnull()) & (df.Lat.notnull()) & (df.Long.notnull())][['STREET', 'Lat', 'Long']]
points_with_street
```

```python
def point2closestStreet(point):
  """
  Returns the closest street to a point
  """
  # Find the closest point in points_with_street to point
  distance_list = np.linalg.norm(points_with_street[['Lat', 'Long']] - point, axis=1)
  closest_point_index = np.argmin(distance_list)
  closest_point = points_with_street.iloc[closest_point_index]
  return closest_point.STREET
```

```python
point2street = {tuple(point):point2closestStreet(point) for point in points_without_street.values}
print(point2street)
```

```python
# Replace missing STREETs in df using point2street
for index, row in df.iterrows():
    if pd.isnull(row['STREET']) and not pd.isnull(row['Lat']) and not pd.isnull(row['Long']):
        if (row['Lat'], row['Long']) in point2street:
            df.loc[index, 'STREET'] = point2street[(row['Lat'], row['Long'])]
```

```python
df.isnull().sum()
```

Even if we do not know how it is defined exactly, we can use the REPORTING_AREA: we take the most common (mode) STREET value for each REPORTING_AREA, and then input it where the street is missing, hopefully finding a few more STREETs we could not find with coordinates only.

```python
def get_street_from_reporting_area(reporting_area):
  """
  Find the most common STREET for a given REPORTING_AREA
  """
  filtered_df = df[df.STREET.notnull()]
  return filtered_df[filtered_df.REPORTING_AREA == reporting_area].STREET.mode()[0]
```

```python
df_reporting_area_but_no_street = df[(df.STREET.isnull()) & (df.REPORTING_AREA.notnull())]
reporting_area2street = {reporting_area:get_street_from_reporting_area(reporting_area) for reporting_area in df_reporting_area_but_no_street.REPORTING_AREA.unique()}
print(reporting_area2street)
```

```python
# We finally use reporting_area2street to replace missing STREET values
for index, row in df.iterrows():
    if pd.isnull(row['STREET']) and not pd.isnull(row['REPORTING_AREA']):
        if row['REPORTING_AREA'] in reporting_area2street:
            df.loc[index, 'STREET'] = reporting_area2street[row['REPORTING_AREA']]
del reporting_area2street # We need to save on memory
```

To summarize, we used STREET to approximate Lat and Long, and in turn we used Lat and Long to find STREET. We also used STREET to guess REPORTING_AREA and DISTRICT. Now we go full circle and guess the REPORTING_AREA from the STREET, using the same techniques as before.

```python
def street2reporting_area(street):
  """
  Find the most common REPORTING_AREA for a given STREET
  """
  filtered_df = df[df.REPORTING_AREA.notnull()]
  mode_values = filtered_df[filtered_df.STREET == street].REPORTING_AREA.mode()
  if mode_values.empty:
    return np.nan
  else:
    return mode_values[0]
```

Unfortunately, RAM requirements are so huge that random forest is simply unusable, even in batches. The same goes for linear regression and classification when we try to guess categorical attributes. So we use MICE on individual chunks of data, and for some numeric data only, so that it hopefully does not crash the session...

```python
missing_columns = ["REPORTING_AREA", "Lat", "Long"]
num_features = ["YEAR", "MONTH", "HOUR", "UCR_PART", "Lat", "Long"]


def iterative_imputation_chunked(df, chunk_size=10000):
    """Performs iterative imputation in chunks to reduce memory usage."""
    imp_data = pd.DataFrame(columns=missing_columns)

    # Scale numeric features before imputation
    scaler = StandardScaler()
    df[num_features] = scaler.fit_transform(df[num_features])

    for i in range(0, df.shape[0], chunk_size):
        chunk = df[i : i + chunk_size].copy()
        mice_impute_pipe = IterativeImputer(
            estimator=KNeighborsRegressor(n_neighbors=10),
            max_iter=10,  # Reduced for faster processing
            initial_strategy="mean",  # or other strategy
            random_state=0,
        )

        imputed_chunk = mice_impute_pipe.fit_transform(
            chunk[["REPORTING_AREA", "YEAR", "MONTH", "HOUR", "UCR_PART", "Lat", "Long"]]
        )
        imputed_chunk = pd.DataFrame(
            imputed_chunk,
            columns=["REPORTING_AREA", "YEAR", "MONTH", "HOUR", "UCR_PART", "Lat", "Long"],
            index=chunk.index,
        )

        # Invert scaling to get original values
        imputed_chunk[num_features] = scaler.inverse_transform(imputed_chunk[num_features])
        imp_data = pd.concat([imp_data, imputed_chunk[missing_columns]])

    return imp_data


# Example usage:
IMP_DATA = iterative_imputation_chunked(df.copy())
print(IMP_DATA)
```

```python
df.update(IMP_DATA)
del IMP_DATA # We need to save on memory
```

We have seen that some categorical elements appear only once or twice in our big dataset.
```python
def encode(original_dataframe, feature_to_encode):
    dummies = pd.get_dummies(original_dataframe[[feature_to_encode]], dummy_na=True)
    res = pd.concat([original_dataframe, dummies], axis=1)
    res = res.drop([feature_to_encode], axis=1)
    return (res)
```

## Classification

Since we have experience using them, we chose to classify with a feedforward neural network (FNN). FNNs are highly customizable, easy to adapt to our dataset, and generally very flexible. We will use an extremely simple architecture to keep training time short.

### Messy dataset

```python
# Calculate the unique target labels and their counts
unique, count = np.unique(y1, return_counts=True)

# Remove rows with classes with < 5 samples
for i in range(len(unique)):
    if count[i] < 5:
        indices = np.where(y1 == unique[i])[0]
        X1 = np.delete(X1, indices, axis=0)
        y1 = np.delete(y1, indices, axis=0)

unique_labels1, count = np.unique(y1, return_counts=True)
print('Target shape', y1.shape)

for u in range(len(unique_labels1)):
    print(f'Class {u}: {unique_labels1[u]} has {count[u]} samples')
```

```python
def class2int(offense):
  return unique.tolist().index(offense)

y1 = np.array([class2int(offense) for offense in y1])
```

```python
ts=3000 # test size

# Split the dataset into a combined training and validation set, and a separate test set
X1_train_val, X1_test, y1_train_val, y1_test = train_test_split(
    X1,
    y1,
    test_size=ts,
    random_state=42,
    stratify=y1
)

# Further split the combined training and validation set into a training set and a validation set
X1_train, X1_val, y1_train, y1_val = train_test_split(
    X1_train_val,
    y1_train_val,
    test_size=ts,
    random_state=42,
    stratify=y1_train_val
)

# Print the shapes of the resulting sets
print('Training set shape:\t', X1_train.shape, y1_train.shape)
print('Validation set shape:\t', X1_val.shape, y1_val.shape)
print('Test set shape:\t\t', X1_test.shape, y1_test.shape)
```

```python
# Apply one-hot encoding to training labels
y1_train = tfk.utils.to_categorical(y1_train, num_classes=len(unique))

# Apply one-hot encoding to validation labels
y1_val = tfk.utils.to_categorical(y1_val, num_classes=len(unique))

# Apply one-hot encoding to test labels
y1_test = tfk.utils.to_categorical(y1_test, num_classes=len(unique))

# Display shapes of the encoded label sets
print('Training set target shape:\t', y1_train.shape)
print('Validation set target shape:\t', y1_val.shape)
print('Test set target shape:\t\t', y1_test.shape)
```

```python
# Determine the input shape, corresponding to the number of features
input_shape = X1_train.shape[1:]
print(f'Input shape of the network {input_shape}')

# Determine the output shape, corresponding to the number of classes
output_shape = y1_train.shape[1]
print(f'Output shape of the network {output_shape}')
```

```python
# Batch size: number of samples processed in each training iteration
batch_size = 16

# Number of epochs: times the entire dataset is passed through the network during training
epochs = 50

# Learning rate: step size for updating the model's weights
learning_rate = 0.001
```

```python
def build_model(
    input_shape=input_shape,
    output_shape=output_shape,
    learning_rate=learning_rate,
    seed=42):

    # Fix randomness
    tf.random.set_seed(seed)

    # Build the neural network layer by layer
    inputs = tfkl.Input(shape=input_shape, name='Input')

    # Add hidden layer with ReLU activation
    x = tfkl.Dense(units=16, name='Hidden')(inputs)
    x = tfkl.Activation('relu', name='HiddenActivation')(x) # ReLU

    # Add output layer with softmax activation
    x = tfkl.Dense(units=output_shape, name='Output')(x)
    outputs = tfkl.Activation('softmax', name='Softmax')(x)

    # Connect input and output through the Model class
    model = tfk.Model(inputs=inputs, outputs=outputs, name='FeedforwardNeuralNetwork')

    # Compile the model with loss, optimizer, and metrics
    loss = tfk.losses.CategoricalCrossentropy()
    optimizer = tfk.optimizers.Adam(learning_rate)
    metrics = ['accuracy']
    model.compile(loss=loss, optimizer=optimizer, metrics=metrics)

    # Return the model
    return model
```

```python
# Build the model with specified input and output shapes
model1 = build_model()

# Display a summary of the model architecture
model1.summary(expand_nested=True, show_trainable=True)

# Plot the model architecture
tfk.utils.plot_model(model1, expand_nested=True, show_trainable=True, show_shapes=True, dpi=70)
```

```python
# Use early stopping to save time
callback = tfk.callbacks.EarlyStopping(monitor='val_accuracy', patience=3, restore_best_weights=True)

# Train the model and store the training history
history = model1.fit(
    x=X1_train,
    y=y1_train,
    batch_size=batch_size,
    epochs=epochs,
    validation_data=(X1_val, y1_val),
    callbacks=[callback]
).history

# Calculate the final validation accuracy
final_val_accuracy = round(history['val_accuracy'][-1] * 100, 2)
```

```python
def confusion_matrix_plot(X_test, y_test, model, unique_labels):
  # Predict class probabilities and get predicted classes
  test_predictions = model.predict(X_test, verbose=0)
  test_predictions = np.argmax(test_predictions, axis=-1)

  # Extract ground truth classes
  test_gt = np.argmax(y_test, axis=-1)

  # Calculate and display test set accuracy
  test_accuracy = accuracy_score(test_gt, test_predictions)
  print(f'Accuracy score over the test set: {round(test_accuracy, 4)}')

  # Calculate and display test set precision
  test_precision = precision_score(test_gt, test_predictions, average='weighted')
  print(f'Precision score (average) over the test set: {round(test_precision, 4)}')

  # Calculate and display test set recall
  test_recall = recall_score(test_gt, test_predictions, average='weighted')
  print(f'Recall score (average) over the test set: {round(test_recall, 4)}')

  # Calculate and display test set F1 score
  test_f1 = f1_score(test_gt, test_predictions, average='weighted')
  print(f'F1 score (average) over the test set: {round(test_f1, 4)}')

  # Compute the confusion matrix
  cm = confusion_matrix(test_gt, test_predictions)

  # Create labels combining confusion matrix values
  labels = np.array([f"{num}" for num in cm.flatten()]).reshape(cm.shape)

  # Plot the confusion matrix with class labels
  plt.figure(figsize=(15, 15))
  sns.heatmap(cm, annot=labels, fmt='', xticklabels=unique_labels, yticklabels=unique_labels, cmap='Blues')
  plt.xlabel('True labels')
  plt.ylabel('Predicted labels')
  plt.show()
```

```python
confusion_matrix_plot(X1_test, y1_test, model1, unique_labels1)
```

```python
# Apply one-hot encoding to training labels
y2_train = tfk.utils.to_categorical(y2_train, num_classes=len(unique))

# Apply one-hot encoding to validation labels
y2_val = tfk.utils.to_categorical(y2_val, num_classes=len(unique))

# Apply one-hot encoding to test labels
y2_test = tfk.utils.to_categorical(y2_test, num_classes=len(unique))

# Display shapes of the encoded label sets
print('Training set target shape:\t', y2_train.shape)
print('Validation set target shape:\t', y2_val.shape)
print('Test set target shape:\t\t', y2_test.shape)
```

```python
# Determine the input shape, corresponding to the number of features
input_shape = X2_train.shape[1:]
print(f'Input shape of the network {input_shape}')

# Determine the output shape, corresponding to the number of classes
output_shape = y2_train.shape[1]
print(f'Output shape of the network {output_shape}')
```

```python
def build_model(
    input_shape=input_shape,
    output_shape=output_shape,
    learning_rate=learning_rate,
    seed=42):

    # Fix randomness
    tf.random.set_seed(seed)

    # Build the neural network layer by layer
    inputs = tfkl.Input(shape=input_shape, name='Input')

    # Add hidden layer with ReLU activation
    x = tfkl.Dense(units=16, name='Hidden')(inputs)
    x = tfkl.Activation('relu', name='HiddenActivation')(x) # ReLU

    # Add output layer with softmax activation
    x = tfkl.Dense(units=output_shape, name='Output')(x)
    outputs = tfkl.Activation('softmax', name='Softmax')(x)

    # Connect input and output through the Model class
    model = tfk.Model(inputs=inputs, outputs=outputs, name='FeedforwardNeuralNetwork')

    # Compile the model with loss, optimizer, and metrics
    loss = tfk.losses.CategoricalCrossentropy()
    optimizer = tfk.optimizers.Adam(learning_rate)
    metrics = ['accuracy']
    model.compile(loss=loss, optimizer=optimizer, metrics=metrics)

    # Return the model
    return model
```

```python
# Build the model with specified input and output shapes
model2 = build_model() # The same as model1
```

```python
# Use early stopping to save time
callback = tfk.callbacks.EarlyStopping(monitor='val_accuracy', patience=3, restore_best_weights=True)

# Train the model and store the training history
history = model2.fit(
    x=X2_train,
    y=y2_train,
    batch_size=batch_size,
    epochs=epochs,
    validation_data=(X2_val, y2_val),
    callbacks=[callback]
).history

# Calculate the final validation accuracy
final_val_accuracy = round(history['val_accuracy'][-1] * 100, 2)
```
